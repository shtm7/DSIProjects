```python
# Importing required libraries.
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline 
sns.set(color_codes=True)
```

```python
df = pd.read_csv("UsedDevices-Haraj-f.csv")
df=df.drop(['Unnamed: 0'],axis=1)
# To display the top 5 rows
df.head(5)
```

```python
# Checking the data type
df.dtypes
```

```python
# Finding the null values.
print(df.isnull().sum())
```

```python
sns.boxplot(x=df['upRank'])
```

```python
sns.boxplot(x=df['commentCount'])
```

```python
df[['commentCount','upRank']].plot.box()
```

```python
# Plotting a Histogram
df.tags.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title('Number of tags by ID')
plt.ylabel('ID')
plt.xlabel('Tages');
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
corr=df.corr()
corr.style.background_gradient(cmap='coolwarm').set_precision(2)
```

```python
df.postDate.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title('Number of postDate by ID')
plt.ylabel('ID')
plt.xlabel('postDate');
```

```python
df.updateDate.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title('Number of updateDate by ID')
plt.ylabel('ID')
plt.xlabel('updateDate');
```
